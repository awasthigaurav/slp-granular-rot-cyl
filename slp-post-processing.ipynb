```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score
from scipy.optimize import curve_fit
```

### shear rate = 0.1

```python
intf = pd.read_csv('intf0-1.dat',header=None,sep='\s')
intf = intf.set_axis(['x','ymin','ymax'],axis=1)
intf
```

```python
x = intf['x']
x = x[1:-1]
ymin = intf['ymin']
ymin = ymin[1:-1]
ymax = intf['ymax']
ymax = ymax[1:-1]
```

```python
plt.plot(x,ymin,marker='.')
plt.plot(x,ymax,marker='.')
delta = np.min(ymin-ymax)*np.sqrt(1-(x**2)/25)
plt.plot(x,delta)
plt.plot(x,ymin-ymax,marker='.')
plt.legend(['ymin','ymax','function','ymin-ymax'])
plt.title('Shear rate = 0.1')
plt.ylabel('y')
plt.xlabel('x')
print(r2_score(delta,ymin-ymax))
```

## Omega vs r and omega vs theta in a 60 degree slice of the cylinder

```python
df = pd.read_csv('avv.dat',header=None,sep='\s')
df.columns = ['x','y','phi','vx','vy','r','omega','theta','vtheta','y-ymin','ymax']
df['theta']*=180/np.pi
df = df[df['r']<5]
df
```

```python
df1 = df.loc[(df['theta']>-120) & (df['theta']<-60)]
df1['theta']+=90
df1 = df1[df1['y-ymin']<0]
df1
```

```python
delta = 0.1
df1['r-chunk'] = np.floor(df1['r']/delta)+1
df1 = df1[df1['y-ymin']<0]
df1
```

```python
for r in np.arange(1,np.max(df1['r-chunk'])+1,1):
    df2 = df1[df1['r-chunk']==r]
    df2 = df2.sort_values(by='theta')
    df2.head()
    plt.plot(df2['theta'],df2['omega'],marker='.')

    for x,y,text in zip(df2['theta'],df2['omega'],df2['r']):
        text = "{:.3f}".format(text)
        plt.annotate(text, # this is the text
                 (x,y), # these are the coordinates to position the label
                #  textcoords="offset points", # how to position the text
                #  xytext=(0,10), # distance from text to points (x,y)
                #  ha='center'# horizontal alignment can be left, right or center
                     )

    plt.xlabel('Theta')
    plt.ylabel('Omega')
    plt.title(f'chunk={r}')
    plt.show()
```

## Curve fitting for
 - Omega vs r
 - vx vs y
 - vy vs y

```python
df = pd.read_csv('avv0-1.dat',header=None,sep='\s')
df.columns = ['x','y','phi','vx','vy','r','omega','theta','vtheta','y-ymin','ymax']
df['theta']*=180/np.pi
df = df[df['r']<=5]
df
```

```python
df1 = df.loc[(df['theta']>-120) & (df['theta']<-60)]
df1['theta']+=90
df1
```

```python
df2 = df1.sort_values(by='r')
df2 = df2[df2['y-ymin']<=0]
plt.plot(df2['r'],np.abs(df2['omega']),marker='.')
plt.xlabel('r')
plt.ylabel('omega')
plt.title(r'$\omega$ vs r for -30<$\theta$<30')
```

```python
# plt.semilogx(df2['r'],np.abs(df2['omega']),marker='.')
# plt.xlabel('r')
# plt.ylabel('omega')
```

```python
def fun(r,a,b,c):
    return a - b*np.exp(-c*r)

# df3 = df2.loc[df2['r']<4.5]
parameters, covariance = curve_fit(fun, df2['r'][:-4], np.abs(df2['omega'][:-4]))

fit_a,fit_b,fit_c=parameters
y = fun(df2['r'][:-4],fit_a,fit_b,fit_c)
plt.plot(df2['r'][:-4],np.abs(df2['omega'][:-4]),marker='.')
plt.plot(df2['r'][:-4],y)
plt.xlabel('r')
plt.ylabel('omega')
print('R2 = ', r2_score(np.abs(df2['omega'][:-4]),y))
print(fit_a,fit_b,fit_c)
```

```python
df3 = df2.loc[df2['r']<4.5]
parameters, covariance = curve_fit(fun, df3['r'], np.abs(df3['omega']))

fit_a,fit_b,fit_c=np.round(parameters,3)
y = fun(df3['r'],fit_a,fit_b,fit_c)
plt.plot(df3['r'],np.abs(df3['omega']),marker='.')
plt.plot(df3['r'],y)
plt.xlabel('r')
plt.ylabel('omega')
print('R2 = ', r2_score(np.abs(df3['omega']),y))
print(fit_a,fit_b,fit_c)
plt.title(r'$\omega$ vs r for -30<$\theta$<30')
plt.legend(['Data',f'{fit_a}-{fit_b}exp(-{fit_c}r)'])
```

```python
delta = 0.1
df3['rchunk'] = np.floor(df3['r']/delta)+1
chunked_omega = df3.groupby('rchunk')['omega'].mean()
plt.plot(np.arange(1.5,4.5,0.1), np.abs(np.asarray(chunked_omega)),marker='.')
plt.plot(df3['r'],y)
plt.ylabel('omega')
plt.xlabel('r')
plt.title(r'$\omega$ vs r (bins of size 0.1)')
plt.legend(['Data',f'{fit_a}-{fit_b}exp(-{fit_c}r)'])
```

```python
values_to_match = [-4.25, -3.25, -2.25, -1.25, -0.25, 0.25, 1.25, 2.25, 3.25, 4.25]
df3 = df[df['x'].isin(values_to_match)]
df3 = df3[(df3['y']<df3['ymax'])]
sns.lineplot(data=df3, x = df3['vx'],y = df3['y-ymin'],hue=df3['x'],legend=values_to_match,palette='coolwarm')
plt.title(r'$y-y_{min}$ vs $v_{x}$')
plt.show()
```

```python
#for fitting vx
def fun2(y,a,b):
    return a*y*(1-np.exp(-b*y))
```

```python
inv_gamma_dot = []
a_values = []
b_values = []
```

```python
xcoord = -4.25
df4 = df3[df3['x']==xcoord]
df4 = df4[df4['y-ymin']>=0]
df_line = df4

vx = df_line['vx'][:,np.newaxis]
m,_,_,_ = np.linalg.lstsq(vx,df_line['y-ymin'])

print(m,'R2=',r2_score(df_line['y-ymin'],m*df_line['vx']))
plt.plot(df4['y-ymin'],df4['vx'],marker='o')
plt.plot(df_line['y-ymin'],1/m*df_line['y-ymin'])

par,cov = curve_fit(fun2,df4['y-ymin'],df4['vx'])
a,b=np.round(par,3)
plt.plot(df4['y-ymin'],fun2(df4['y-ymin'],a,b))
plt.xlabel('y-ymin')
plt.ylabel('vx')
plt.title(r'$v_{x}$  vs $y-y_{min}$ at x=-4.25')
plt.legend(['Data',f'vx={np.round(1/m,3)}(y-ymin)',f'vx={a}(y-ymin)(1-exp(-{b}(y-ymin)))'])
print(a,b,'R2=',r2_score(df4['vx'],fun2(df4['y-ymin'],a,b)))

inv_gamma_dot.append(m)
a_values.append(a)
b_values.append(b)
```
